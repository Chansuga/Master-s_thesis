```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[-1,1,1],[1,1,1],[1,1,1]])

plt.imshow(A, cmap='gray', interpolation='nearest')
plt.title("Spin Configuration")
plt.show()

if np.sum(A==1) >= 5:
    A = A
else:
    A = A*-1

plt.imshow(A, cmap='gray', interpolation='nearest')
plt.title("Spin Configuration")
plt.show()
```
