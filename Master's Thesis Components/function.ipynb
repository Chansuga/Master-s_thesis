```python
import numpy as np
import matplotlib.pyplot as plt

# x軸の値を生成
x = np.linspace(0, 3, 100)  # 0から3までの区間を100等分

# サイン関数とコサイン関数の値を計算
y = np.sinh(x) / (np.cosh(x) + np.exp(-1))

# グラフの設定
plt.figure(figsize=(8, 5))
# plt.grid(True)

# 関数をプロット
plt.plot(x, y, color='black')

# y=0の点線を追加
plt.axhline(0, color='black', linestyle=':')

# 凡例を追加
plt.legend()

# x軸とy軸にラベルを追加
plt.xlabel(r'2$\beta$H')
plt.ylabel(r'$\langle M \rangle$')

# x軸の範囲を調整して原点を左端に配置
plt.xlim(0, 3)

# グラフを画像として保存
plt.savefig('graph/fukushima(3_140).png')
```
