```python
import numpy as np
import matplotlib.pyplot as plt

# xの値の範囲を設定
x = np.linspace(-2, 2, 400)

# 関数を計算
y = (1/2) * np.log(np.cosh(2 * x))

# グラフを描画
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = (1/2) * ln(cosh(2x))')
plt.xlabel('$K$')
plt.ylabel('$K^{,}$')
plt.legend()
plt.title('Graph of y = (1/2) * ln(cosh(2x))')
plt.grid(True)
plt.show()
```
