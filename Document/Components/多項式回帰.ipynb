sklearnによる多項式回帰

```python
import numpy as np
import sklearn.linear_model as lm
import matplotlib.pyplot as plt

data_size=11
#0~1までを11個区切りで表す。
x=np.linspace(0,1,data_size)
y=np.array([3.4, 2.0, 1.6, 1.6, 2.5, 2.8, 2.6, 2.3, 2.1, 2.4, 3.8])


# 単純な多項式近似
lrp = lm.LinearRegression()

for deg in [2,8,20]: # 最大べき
    power_matrix_x = np.vander(x, deg+1) # 計画行列の作成
    lrp.fit(power_matrix_x, y)
    # 予測
    x_lrp = np.linspace(0., 1., 100)
    y_lrp = lrp.predict(np.vander(x_lrp, deg+1))
    # 近似曲線の描画
    plt.plot(x_lrp, y_lrp)
    # データの描画
    plt.plot(x, y, "ob", ms=8)
    plt.ylim(1.0,4.0)
```

```python
deg=2
power_matrix_x = np.vander(x, deg+1) # 計画行列の作成
lrp.fit(power_matrix_x, y)
# 予測
x_lrp = np.linspace(0., 1., 100)
y_lrp = lrp.predict(np.vander(x_lrp, deg+1))
# 近似曲線の描画
plt.plot(x_lrp, y_lrp)
# データの描画
plt.plot(x, y, "ob", ms=5)
plt.ylim(1.0,4.0)
plt.savefig("./graph/多項式回帰(a).png")
```
